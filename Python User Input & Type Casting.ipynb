User input and Casting

what is User Input?
User input is data entered by the user during program execution.

What is Type Casting?
Type Casting is converting one data type into another.

Why is Type Casting uded?
It is used to perform operations on the correct data type.

examples:
int() converts a value to an integer.
float() convert a value to a decimal number.
str() converts a value to a string.

Question1:
Write a programme to take the users name and age as input and display their sum.

```python
name=input("Enter your name:")
age=int(input("Enter your age:"))
print("Name:", name)
print("Age:", age)
```

Write a programme to take two integers from the user and display their sum.

```python
a=int(input("Enter your first number:"))
b=int(input("Enter your second number:"))
sum=a+b
print("Enter the first number", a)
print("Enther the second number:", b)
print("sum:", sum)
```

Write a program to take the length and width of a rectangle as input and calculate its area.

```python
length=float(input("length:"))
width=float(input("width"))
area=length*width
print("Area:", area)
```
