if_else with Boolean values:

If statement:Executes code when a condition is True.
Else Statement: Executes code when a condition is False.
Boolean Values: True and False are Boolean values used for decision_making.

```python
Question1: Write a program that checks whether a student has passed.
```

```python
student_has_passed=True

if(student_has_passed):
    print("student passed")
else:
    print("Student failed")
```
