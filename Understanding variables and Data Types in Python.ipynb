Variables:
A variable is named storage location used to store data in program.
Data Types:
A data type defines the kind of value a variable can hold, such as integer, float, string, or boolean.


Question1:
create a variable and give a input from users,
print their values and data types

```python
name=input()
score=int(input())
subject=input()
print("Name:", name)
print("Score:", score)
print("subject:", subject)

print(type(name))
print(type(score))
print(type(subject))
```
